```python
""" This function calculate the information gain for only one feature """
def information_gain(feature , y):
    counterForRepublican = 0
    counterForDemocrat = 0
    for i in range(len(feature)):
        if y[i] == 'democrat':
            counterForDemocrat = counterForDemocrat + 1
        else:
            counterForRepublican = counterForRepublican + 1
    
    counterForYesAndDemocrate = 0 
    counterForYesAndRepublican = 0
    
    for i in range(len(feature)):
        if feature[i] == 'y' and y[i] == 'democrat' :
            counterForYesAndDemocrate = counterForYesAndDemocrate + 1
        if feature[i] == 'y' and y[i] == 'republican' :
            counterForYesAndRepublican = counterForYesAndRepublican + 1
            
    
    counterForNoAndDemocrate = 0 
    counterForNoAndRepublican = 0
    
    for i in range(len(feature)):
        if feature[i] == 'n' and y[i] == 'democrat' :
            counterForNoAndDemocrate = counterForNoAndDemocrate + 1
        if feature[i] == 'n' and y[i] == 'republican' :
            counterForNoAndRepublican = counterForNoAndRepublican + 1
        
    
    """ H(S) the entropy for the parent """
    Hs = (-counterForRepublican/len(feature) * np.log(counterForRepublican/len(feature))) - (counterForDemocrat/len(feature) * np.log(counterForDemocrat/len(feature)))
    
    HsYes = (-counterForYesAndDemocrate/(counterForYesAndDemocrate + counterForYesAndRepublican) * np.log(counterForYesAndDemocrate/(counterForYesAndDemocrate + counterForYesAndRepublican))) - (counterForYesAndRepublican/(counterForYesAndDemocrate + counterForYesAndRepublican) * np.log(counterForYesAndRepublican/(counterForYesAndDemocrate + counterForYesAndRepublican)))
    
    HsNo = (-counterForNoAndDemocrate/(counterForNoAndDemocrate + counterForNoAndRepublican) * np.log(counterForNoAndDemocrate/(counterForNoAndDemocrate + counterForNoAndRepublican))) - (counterForNoAndRepublican/(counterForNoAndDemocrate + counterForNoAndRepublican) * np.log(counterForNoAndRepublican/(counterForNoAndDemocrate + counterForNoAndRepublican)))
    
    gain = Hs - HsYes - HsNo
    
    return gain
    
```
