```python
""" This function calculate the information gain for only one feature """
def information_gain(featureDataFrame , colIndex , y_DataFrame):
    
    feature_array = featureDataFrame.to_numpy()
    #print(feature_array)
    feature = (feature_array[: , colIndex]).reshape(len(featureDataFrame) , 1)
    
    #print(feature)
    #print(feature.shape)
    
    #print("***********************************************")
    
    y = (y_DataFrame.to_numpy()).reshape(len(y_DataFrame) , 1)
    #print(y)
    
    H_parent = 0
    H_child1 = 0
    H_child2 = 0
    
    counter1 = 0 #To count how many number of democrat
    counter2 = 0 #To count how many number of republican
    for i in range(len(y)):
        if y[i] == "republican" :
            counter1 = counter1 + 1
        else:
            counter2 = counter2 + 1
            
    
            
    counterForYes1 = 0 #To count how many number of republican when the feature is yes
    counterForYes2 = 0 #To count how many number of democrat when the feature is yes
    for i in range(len(y)):
        if y[i] == "republican" and feature[i] == 'y':
            counterForYes1 = counterForYes1 + 1
        elif y[i] == "democrat" and feature[i] == 'y':
            counterForYes2 = counterForYes2 + 1
            
    if counterForYes1 == 0:
        H_child1 = (- ( counterForYes2 / (counterForYes1 + counterForYes2) * np.log2(counterForYes2 / (counterForYes1 + counterForYes2))))
    elif counterForYes2 == 0:
        H_child1 = (( -counterForYes1 / (counterForYes1 + counterForYes2) * np.log2(counterForYes1 / (counterForYes1 + counterForYes2))) )
    elif counterForYes1 == 0 and counterForYes2 == 0:
        H_child1 = 0
    else:
        H_child1 = (( -counterForYes1 / (counterForYes1 + counterForYes2) * np.log2(counterForYes1 / (counterForYes1 + counterForYes2))) - ( counterForYes2 / (counterForYes1 + counterForYes2) * np.log2(counterForYes2 / (counterForYes1 + counterForYes2))))
    
    
    counterForNo1 = 0 #To count how many number of republican when the feature is No
    counterForNo2 = 0 #To count how many number of democrat when the feature is No
    for i in range(len(y)):
        if y[i] == "republican" and feature[i] == 'n':
            counterForNo1 = counterForNo1 + 1
        elif y[i] == "democrat" and feature[i] == 'n':
            counterForNo2 = counterForNo2 + 1
            
    
    if counterForNo1 == 0:
        H_child2 = (- ( counterForNo2 / (counterForNo1 + counterForNo2) * np.log2(counterForNo2 / (counterForNo1 + counterForNo2))))
    elif counterForNo2 == 0:
         H_child2 = (( -counterForNo1 / (counterForNo1 + counterForNo2) * np.log2(counterForNo1 / (counterForNo1 + counterForNo2))) )
    elif counterForNo1 == 0 and counterForNo2 == 0:
        H_child2 = 0
    else:
         H_child2 = (( -counterForNo1 / (counterForNo1 + counterForNo2) * np.log2(counterForNo1 / (counterForNo1 + counterForNo2))) - ( counterForNo2 / (counterForNo1 + counterForNo2) * np.log2(counterForNo2 / (counterForNo1 + counterForNo2))))
    
    
    H_parent = (( -counter1 / (counter1 + counter2) * np.log2(counter1 / (counter1 + counter2))) - ( counter2 / (counter1 + counter2) * np.log2(counter2 / (counter1 + counter2))))
            
    
   
    gain = (float(H_parent) - float(H_child1) - float(H_child2))
    
    return gain
    
```

```python
class Node:
    def __init__(self):
        self.left = None
        self.right = None
        self.attribute = ""
        self.value = ""
        self.classDemocrat = 0
        self.classRepublican = 0
        self.label = None
    
    def addNode(self , attribute , value , classRepublican = 0 , classDemocrat = 0 ):
        self.attribute = attribute
        self.classDemocrat = classDemocrat
        self.classRepublican = classRepublican
        self.value = value
```

```python
class DecisionTree:
    def __init__(self):
        self.root = Node()
        self.root.addNode('temp' , "temp")
        
    def createDecisionTree(self , X_DataFrame , y_DataFrame , tree):
        
        X = X_DataFrame.to_numpy()
        #print(feature_array)
        
        y = (y_DataFrame.to_numpy()).reshape(len(y_DataFrame) , 1)
        
        total = X.shape[0]
        
        ones = 0 #To count how many number of democrat
        zeros = 0 #To count how many number of republican
        for i in range(len(y)):
            if y[i] == "republican" :
                ones = ones + 1
            else:
                zeros = zeros + 1
        
        if X.shape[1] == 1 or total == ones or total == zeros:
            if zeros >= ones:
                tree.label = 'democrat'
            else:
                tree.label = 'republican'
            return   
        
        else:
            bestFeature , index = bestAttribute(X_DataFrame , y_DataFrame) #it will hold the name of the feature
            tree.left = Node()
            tree.right = Node()
            
            """ Count the number of reoublican with the only exisiting of the bestFeature"""
            
            
            republicanCountYes = 0
            democratCountYes = 0
            for i in range(len(X)):
                if y[i] == "republican" and X[i , index] == 'y':
                    republicanCountYes = republicanCountYes + 1
                elif y[i] == "democrat" and X[i , index] == 'y':
                    democratCountYes = democratCountYes + 1 
            
            
            republicanCountNo = 0
            democratCountNo = 0
            for i in range(len(X)):
                if y[i] == "republican" and X[i , index] == 'n':
                    republicanCountNo = republicanCountNo + 1
                elif y[i] == "democrat" and X[i , index] == 'n':
                    democratCountNo = democratCountNo + 1 
                    
            tree.left.addNode(bestFeature , "y" , republicanCountYes , democratCountYes)
            tree.right.addNode(bestFeature , "n" , republicanCountNo , democratCountNo)
            
            self.createDecisionTree(X_DataFrame[X_DataFrame[bestFeature]==0].drop([bestFeature], axis=1) , y_DataFrame , tree.left)
            self.createDecisionTree(X_DataFrame[X_DataFrame[bestFeature]==0].drop([bestFeature], axis=1) , y_DataFrame , tree.right)
            
    def predict(self , data , root):
        if root.label is not None:
            return root.label
        elif data[root.left.attribute][data.index.tolist()[0]] == 1:
            return self.predict(data, root.right)
        else:
            return self.predict(data, root.left)
        
    def countNodes(self,node):
        if(node.left is not None and node.right is not None):
            return 2 + self.countNodes(node.left) + self.countNodes(node.right)
        return 0

    
    def countLeaf(self,node):
        if(node.left is None and node.right is None):
            return 1
        return self.countLeaf(node.left) + self.countLeaf(node.right)
        
    

    
def calculateAccuracy(data , label , tree):
    y = (label.to_numpy()).reshape(len(label) , 1)
    correctCount = 0
    for i in data.index:
        val = tree.predict(data.iloc[i:i+1, :] , tree.root)
        if val == y[i]:
            correctCount = correctCount + 1
    return correctCount/data.shape[0]*100
```

## Part 1
----------------------------------------------------------------------------------

Use the voting data to train a decision tree to predict political party (Democrat or
Republican) based on the voting record. Use 25% of the members of congress for 
training and the rest for testing. Rerun this experiment five times and notice the 
impact of different random splits of the data into training and test sets. Report the 
sizes and accuracies of these trees in each experiment.

```python
for i in range(5):
    data = dataset.sample(frac = 0.25)
    X = data[col_names]
    y = data[['Class Name']]
    indeces = len(X)
    splitIndex = int(np.ceil(0.25 * indeces)) 
    
    X_train = X.iloc[0:splitIndex , :].values
    X_train = pd.DataFrame(X_train , columns = col_names)
    
    y_train = y.iloc[0:splitIndex , :].values
    y_train = pd.DataFrame(y_train , columns = ['Class Name'])
    
    X_test = X.iloc[splitIndex: , :].values
    X_test = pd.DataFrame(X_test , columns = col_names)
    
    y_test = y.iloc[splitIndex: , :].values
    y_test = pd.DataFrame(y_test , columns = ['Class Name'])
    
    model = DecisionTree()
    model.createDecisionTree(X_train , y_train , model.root)
    
    accuracy = calculateAccuracy(X_test , y_test , model)
    print('Accuracy without handling overfiiting and underfitting = %.2f'%accuracy , '% for iteration ' , i)
    print("Total number of nodes in the tree = "+ str(model.countNodes(model.root)) , ' for iteration ' , i)
    print("***************************************************************")
```

## Part 2
----------------------------------------------------------------------------------

Measure the impact of training set size on the accuracy and the size of the learned 
tree. Consider training set sizes in the range (30-70%). Because of the high variance
due to random splits repeat the experiment with five different random seeds for each
training set size then report the mean, maximum and minimum accuracies at each
training set size. Also measure the mean, max and min tree size.
Start with training data size 30%, 40 Until you reach 70%.
The data set contained many missing values , i.e., votes in which a member of
congress failed to participate. To solve those issue insert for each absent
vote the voting decision of the majority.

```python
splitSize = [0.30 , 0.40 , 0.50 , 0.60 , 0.70]


for i in range(5):
    accuracies = []
    treeSize = []
    print("TrainSize- " , splitSize[i] , " :")
    for j in range(5):
        data = dataset.sample(frac = splitSize[i])
        X = data[col_names]
        y = data[['Class Name']]
        indeces = len(X)
        splitIndex = int(np.ceil(splitSize[i] * indeces)) 

        X_train = X.iloc[0:splitIndex , :].values
        X_train = pd.DataFrame(X_train , columns = col_names)

        y_train = y.iloc[0:splitIndex , :].values
        y_train = pd.DataFrame(y_train , columns = ['Class Name'])

        X_test = X.iloc[splitIndex: , :].values
        X_test = pd.DataFrame(X_test , columns = col_names)

        y_test = y.iloc[splitIndex: , :].values
        y_test = pd.DataFrame(y_test , columns = ['Class Name'])

        model = DecisionTree()
        model.createDecisionTree(X_train , y_train , model.root)

        accuracy = calculateAccuracy(X_test , y_test , model)
        accuracies.append(accuracy)
        treeSize.append(model.countNodes(model.root))
        
        
        print('Accuracy = %.2f'%accuracy , '% for iteration ' , j)
        print("Total number of nodes in the tree = "+ str(model.countNodes(model.root)) , ' for iteration ' , j)
    
    
    print("Report with the min , max ,and mean accuarcies at training size =  " , splitSize[i] * 100 , " %")
    max_accuracy = max(accuracies)
    max_index = accuracies.index(max_accuracy)
    print("Maximum accuracy = " , max_accuracy , " with training size = " , splitSize[i] * 100 , " %")

    print("*******************************************************")

    min_accuracy = min(accuracies)
    min_index = accuracies.index(min_accuracy)
    print("Minimum accuracy = " , min_accuracy , " with training size = " , splitSize[i] * 100 , " %")

    print("*******************************************************")

    total_accuracy = sum(accuracies)
    mean = total_accuracy / len(splitSize)
    print("Mean Accuracies = " , mean , " with training size = " , splitSize[i] * 100 , " %")
    
    
    
    print("Report with the min , max ,and mean sizes at training size =  " , splitSize[i] * 100 , " %")
    max_size = max(treeSize)
    max_index = treeSize.index(max_size)
    print("Maximum tree size = " , max_size , " with training size = " , splitSize[i] * 100 , " %")

    print("*******************************************************")

    min_size = min(treeSize)
    min_index = treeSize.index(min_size)
    print("Minimum tree size = " , min_size , " with training size = " , splitSize[i] * 100 , " %")

    print("*******************************************************")

    total_size = sum(treeSize)
    mean = total_size / len(splitSize)
    print("Mean Sizes = " , mean , " with training size = " , splitSize[i] * 100 , " %")
    print("***************************************************************")
```
